```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('paper', font_scale=1.5)
```

```python
from astropy.visualization import hist
```

```python
df = pd.read_csv('distances.csv.gz')
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6,6), sharex=True, squeeze=True)

_ = hist(df.ratio_disk_size, bins='scott', density=True, histtype='step', label='size = major_axis', ax=ax[0], lw=2)
_ = hist(df.ratio_disk_minor_size, bins='scott', density=True, histtype='step', label='size = minor_axis', ax=ax[0], lw=2, ls='dashed')
ax[0].set_title('Disk')
_ = hist(df.ratio_bulge_size, bins='scott', density=True, histtype='step', label='size = major_axis', ax=ax[1], lw=2)
_ = hist(df.ratio_bulge_minor_size, bins='scott', density=True, histtype='step', label='size = minor_axis', ax=ax[1], lw=2, ls='dashed')
ax[1].set_title('Bulge')
ax[1].set_xlabel('Host SN sep / galaxy size')
ax[0].set_ylabel('PDF')
ax[1].set_ylabel('PDF')
plt.legend(loc='best')
ax[0].set_xlim(0, 20)
ax[1].set_xlim(0, 20)
fig.set_tight_layout(True)
fig.savefig('sn_host_separation_in_units_of_galaxy_size_sqfig.pdf')
```
